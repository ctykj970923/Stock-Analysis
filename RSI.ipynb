```python
def rsi_signal(stockNum):
    ticker = stockNum
    ticker = ticker.strip('\n')
    data = pdr.get_data_yahoo(ticker, start, end)
    data["rsi14"]=ta.RSI(data['Close'], timeperiod=14)
    rsi14=data.iloc[-1].at['rsi14']
    if(60<rsi14<70):
        return True
    else:
        return False
```

```python
##stock have expma signal
newList=[]
with open ('xxxx.HK.txt','r') as f:
    f_contents = f.readlines()
    newList = f_contents
file = open("rsi signal.txt","w")
for stockNum in newList:
    print(stockNum)
    try:
        if(rsi_signal(stockNum))== True:
            print(stockNum)
            file.write(stockNum)
        else:
            continue
    except IndexError:
        continue
    except KeyError:
        continue
file.close()
```
